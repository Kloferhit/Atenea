```python
import numpy as np
import pandas as pd
```

```python
#Fuentes Información

df= pd.read_excel("C:/Renato/script_atenea/Split/SI.xlsx"
,sheet_name="Hoja1")

df_forecast= pd.read_excel("C:/Renato/script_atenea/Split/Output_x.xlsx"
,sheet_name="Hoja1")

df_demanda= pd.read_excel("C:/Renato/script_atenea/Split/IBP.xlsx"
,sheet_name="Hoja1")
```

```python
#DICCIONARIOS y LISTAS

dicc_categoría=df[["Cat2","CATEGORIA_AFO"]].drop_duplicates()
dicc_categoría

familias =df['FAMILIA_AFO'].unique().tolist()

Canales =df['CANAL'].unique().tolist()

df_fam_canales=df[['FAMILIA_AFO','CANAL']].drop_duplicates()

df_dicc_Material=df[["MATERIAL_AFO","FAMILIA_AFO","CATEGORIA_AFO","NEGOCIO_AFO"
                            ]].drop_duplicates()

df_dicc_familia_Cat= df[["FAMILIA_AFO","CATEGORIA_AFO"]
                            ].drop_duplicates()
df_dicc_familia_Cat=pd.DataFrame(df_dicc_familia_Cat)
df_dicc_familia_Cat=df_dicc_familia_Cat.reset_index().drop(columns=["index"])
df_dicc_familia_Cat
```

```python
# DATAFRAME SI average ultimos 3

df=df.drop(columns=["Categoría","Cat2","Oficina de ventas","OF2","Negocio","NEG2","Familia","FAM2","Cliente actual","CA2",
                    "Territorio","TER2","Material histórico","MAT2"])

N = 3

df_num = df.select_dtypes(include=[np.float64])
count= df_num.shape[1]

avg = df.iloc[:,-N:]
avg = avg.mean(axis=1)
df_avg=df.iloc[:,:-count]
df_avg[f"average {N}"]= avg

df_avg
```

```python
#SACANDO DF por cada familia
df_subset={}
df_subset_perc={}
df_fam_canales=df[['FAMILIA_AFO','CANAL']].drop_duplicates()
df_fam_canales["MODERNO"]=0
df_fam_canales["TRADICIONAL"]=0
df_fam_canales=df_fam_canales.reset_index()
df_fam_canales=df_fam_canales.drop(columns=["index"])


for can in Canales:
    for fam in familias:
        df_subset[str(fam)+"_"+str(can)]= df_avg.loc[(df_avg["FAMILIA_AFO"]==fam) & (df_avg["CANAL"]==can)]
        total =df_subset[str(fam)+"_"+str(can)].groupby(by=["CANAL","OFICINA_AFO","CLIENTE_AFO","TERRITORIO_AFO","MATERIAL_AFO"
                                ])[f"average {N}"].sum().sum()
        df_subset_perc[str(fam)+"_"+str(can)]= df_subset[str(fam)+"_"+str(can)].groupby(by=["CANAL","OFICINA_AFO","CLIENTE_AFO","TERRITORIO_AFO","MATERIAL_AFO"
                                ])[f"average {N}"].sum()/total
        index=df_fam_canales.index[(df_fam_canales["FAMILIA_AFO"]==fam) & (df_fam_canales["CANAL"]==can)]

        df_fam_canales.loc[index,can]= df_subset_perc[str(fam)+"_"+str(can)].sum()
        
df_fam_canales=df_fam_canales.groupby(by=["FAMILIA_AFO"],as_index=False
                            )[["MODERNO","TRADICIONAL"]].sum().round()



```

```python
df_fam_canales



```

```python
#ATENEA FORECAST
df_forecast["CANAL_CAT"]=df_forecast["Canal"]+"_"+df_forecast["Categoría"]
df_forecast_best=df_forecast.groupby(by=["CANAL_CAT"]
                            )["Score"].min()
df_forecast_best=df_forecast_best.reset_index()

df_forecast_best_models = df_forecast[(df_forecast["CANAL_CAT"].isin(df_forecast_best["CANAL_CAT"])) &
                               (df_forecast["Score"].isin(df_forecast_best["Score"])) ]


df_forecast_best_models=df_forecast_best_models.merge(dicc_categoría,left_on="Categoría",right_on="Cat2")
df_forecast_best_models=df_forecast_best_models.drop(columns=["Categoría","Cat2"])

df_forecast_best_models_Categoria= df_forecast_best_models.groupby(by=["CATEGORIA_AFO"],as_index=False
                            )["predicción_mes_siguiente"].sum()
df_forecast_best_models_Categoria.rename(columns={'predicción_mes_siguiente':'predicción total cat'},inplace=True)

df_forecast_best_models_perc= df_forecast_best_models.merge(df_forecast_best_models_Categoria,right_on="CATEGORIA_AFO",left_on="CATEGORIA_AFO")

df_forecast_best_models_perc["Porcentaje"]= df_forecast_best_models_perc["predicción_mes_siguiente"]/df_forecast_best_models_perc["predicción total cat"]
df_forecast_best_models_perc
df_forecast_best_models_perc=df_forecast_best_models_perc.drop(columns=["predicción_mes_siguiente","predicción total cat","Modelo","Score","CANAL_CAT"])

df_forecast_best_models_perc
```

```python
#IBP(Demanda)

df_demanda.rename(columns={'Familia Desc.':'FAMILIA_AFO'},inplace=True)

columnas_demanda=df_demanda.columns
columna_predicción_demanda= columnas_demanda[-1]

df_demanda["FAMILIA_AFO"]=df_demanda["FAMILIA_AFO"].replace({"-":" "},regex=True)

df_demanda_familia = df_demanda.groupby(by=["FAMILIA_AFO"],as_index=False
                            )[columna_predicción_demanda].sum()
df_demanda_familia
```

```python
#CREACIÓN DATASET SPLIT FINAL
df_split= df_demanda_familia.merge(df_dicc_familia_Cat,right_on="FAMILIA_AFO",left_on="FAMILIA_AFO")

df_split= df_split.merge(df_forecast_best_models_perc,right_on="CATEGORIA_AFO",left_on="CATEGORIA_AFO")

df_split["Fam_Can"]= df_split["FAMILIA_AFO"]+"_"+df_split["Canal"]
df_split["Split total"]=df_split["Porcentaje"]*df_split[columna_predicción_demanda]


familias_ibp= pd.unique(df_split["FAMILIA_AFO"]).tolist()
canales_ibp= pd.unique(df_split["Canal"]).tolist()


df_final=pd.DataFrame()
for can in canales_ibp:
    for fam in familias_ibp:
        df_fam_subset_escogido=df_fam_canales[df_fam_canales["FAMILIA_AFO"]==fam]
        #print(df_fam_subset_escogido)
        df_subset_escogido=df_subset_perc[str(fam)+"_"+str(can)].reset_index()
        #print(df_subset_escogido)
        #print(df_subset_escogido[f"average {N}"].sum())

        if (df_fam_subset_escogido["MODERNO"].values) + (df_fam_subset_escogido["TRADICIONAL"].values) == 2:
            df_x= df_split[df_split["Fam_Can"]==str(fam)+"_"+str(can)]
            total_fam_canal=df_x["Split total"].values
            df_subset_escogido[f"average {N}"]=df_subset_escogido[f"average {N}"]* total_fam_canal
            df_final=pd.concat([df_final,df_subset_escogido],ignore_index=True)

        elif (df_fam_subset_escogido["MODERNO"].values) + (df_fam_subset_escogido["TRADICIONAL"].values) ==1:
            df_x= df_split[df_split["Fam_Can"]==str(fam)+"_"+str(can)]
            total_fam_canal=df_x[columna_predicción_demanda].values
            df_subset_escogido[f"average {N}"]=df_subset_escogido[f"average {N}"]* total_fam_canal
            df_final=pd.concat([df_final,df_subset_escogido],ignore_index=True)
        else:
            continue

df_final = df_final.merge(df_dicc_Material,right_on="MATERIAL_AFO",left_on="MATERIAL_AFO")

print(df_final.groupby(by=["FAMILIA_AFO"],as_index=False
                            )[f"average {N}"].sum())
print(df_final["average 3"].sum())
```
